```python
# Función para el método del codo
def elbow_method(X):
    inertia_dict = {}
    for k in range(1, 11):
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
        kmeans.fit(X)
        inertia_dict[k] = kmeans.inertia_
    
    # Graficar el codo
    plt.plot(range(1, 11), list(inertia_dict.values()), marker='o', color='b')
    plt.title('Método del Codo')
    plt.xlabel('Número de clusters (k)')
    plt.ylabel('Inercia (Distancia intra-clases)')
    plt.xticks(range(1, 11))
    plt.grid(True)
    plt.show()
    
    return inertia_dict

```

```python
# Función para calcular el índice Davies-Bouldin
def davies_bouldin(X, k):
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    kmeans.fit(X)
    db_index = davies_bouldin_score(X, kmeans.labels_)
    return db_index

# Función para calcular el índice Silhouette
def silhouette_index(X, k):
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    kmeans.fit(X)
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    return silhouette_avg

```

```python
# Función principal que obtiene el reporte de métricas y visualizaciones
def clustering_report(X):
    # Normalizar los datos
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # 1. Método del Codo
    print("Ejecutando el Método del Codo...")
    elbow_dict = elbow_method(X_scaled)
    
    # 2. Davies-Bouldin para k = 2 a 10
    print("Evaluando Davies-Bouldin para k = 2 a 10...")
    db_scores = {k: davies_bouldin(X_scaled, k) for k in range(2, 11)}
    print("Davies-Bouldin Scores:", db_scores)
    
    # 3. Silhouette para k = 2 a 10
    print("Evaluando Silhouette para k = 2 a 10...")
    silhouette_scores = {k: silhouette_index(X_scaled, k) for k in range(2, 11)}
    print("Silhouette Scores:", silhouette_scores)
    
    # Graficar Davies-Bouldin
    plt.plot(list(db_scores.keys()), list(db_scores.values()), marker='o', color='r')
    plt.title('Davies-Bouldin Index')
    plt.xlabel('Número de clusters (k)')
    plt.ylabel('Índice de Davies-Bouldin')
    plt.grid(True)
    plt.show()

    # Graficar Silhouette
    plt.plot(list(silhouette_scores.keys()), list(silhouette_scores.values()), marker='o', color='g')
    plt.title('Índice de Silhouette')
    plt.xlabel('Número de clusters (k)')
    plt.ylabel('Índice de Silhouette')
    plt.grid(True)
    plt.show()

```

```python
class KMeansCustom:
    def __init__(self, max_iter=300, tol=1e-4, random_state=42):
        self.max_iter = max_iter  # Número máximo de iteraciones
        self.tol = tol  # Tolerancia para la convergencia
        self.random_state = random_state  # Semilla aleatoria para la reproducibilidad

    def initialize_centroids(self, X, k):
        """Inicializa los centroides aleatoriamente seleccionando puntos de los datos."""
        np.random.seed(self.random_state)
        random_indices = np.random.choice(X.shape[0], size=k, replace=False)
        centroids = X[random_indices]
        return centroids
    
    def assign_clusters(self, X, centroids):
        """Asigna cada punto al centroide más cercano."""
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)  # Distancia Euclidiana
        labels = np.argmin(distances, axis=1)  # Indices de los centroides más cercanos
        return labels
    
    def update_centroids(self, X, labels, k):
        """Actualiza los centroides calculando la media de los puntos asignados a cada cluster."""
        centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        return centroids
    
    def has_converged(self, old_centroids, centroids):
        """Verifica si los centroides han dejado de cambiar."""
        return np.all(np.abs(old_centroids - centroids) < self.tol)
    
    def fit(self, df, k):
        """Ajusta el algoritmo k-means a la base de datos y devuelve las etiquetas de los clusters."""
        # Preprocesamiento: Estandarizar los datos
        scaler = StandardScaler()
        X = scaler.fit_transform(df)  # Normalizamos las características
        
        # Inicialización de los centroides
        centroids = self.initialize_centroids(X, k)
        labels = np.zeros(X.shape[0])
        
        # Iterar hasta que el algoritmo converja o alcance el número máximo de iteraciones
        for i in range(self.max_iter):
            # Asignar cada punto al cluster más cercano
            new_labels = self.assign_clusters(X, centroids)
            
            # Verificar la convergencia
            if np.array_equal(new_labels, labels):
                print(f'Convergencia alcanzada en la iteración {i}')
                break
            labels = new_labels
            
            # Actualizar los centroides
            old_centroids = centroids
            centroids = self.update_centroids(X, labels, k)
            
            # Comprobar si los centroides han convergido
            if self.has_converged(old_centroids, centroids):
                print(f'Convergencia alcanzada en la iteración {i}')
                break
        
        # Calcular la distancia intra-cluster (inercia)
        intra_cluster_distance = np.sum(np.min(np.linalg.norm(X[:, np.newaxis] - centroids, axis=2), axis=1))
        
        # Calcular la distancia entre clusters
        between_cluster_distance = np.sum([np.linalg.norm(centroids[i] - centroids[j]) 
                                           for i in range(k) for j in range(i+1, k)])
        
        # Almacenar los resultados en los atributos de la clase
        self.centroids = centroids
        self.labels = labels
        self.intra_cluster_distance = intra_cluster_distance
        self.between_cluster_distance = between_cluster_distance
        
        return labels

    def plot_clusters(self, df, k):
        """Genera un gráfico de dispersión para los clusters asignados."""
        plt.figure(figsize=(8, 6))
        plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=self.labels, cmap='viridis', s=50)
        plt.scatter(self.centroids[:, 0], self.centroids[:, 1], c='red', s=200, marker='X')  # Centroides en rojo
        plt.title(f'Clusters generados por KMeans (k={k})')
        plt.xlabel(df.columns[0])
        plt.ylabel(df.columns[1])
        plt.show()

    def get_metrics(self):
        """Devuelve las métricas calculadas (distancia intra-cluster, entre clusters)."""
        return {
            'Intra-cluster Distance (Inercia)': self.intra_cluster_distance,
            'Between-cluster Distance': self.between_cluster_distance,
            'Centroids': self.centroids,
            'Labels': self.labels
        }
```

```python
# Función para entrenar el modelo con los datos de 2024
def train_and_save_kmeans(df_2024, k, model_filename="kmeans_model.pkl"):
    """
    Entrena el modelo KMeans utilizando los datos de 2024 y guarda los centroides.
    df_2024: DataFrame con los datos de los clientes de 2024.
    k: Número de clusters.
    model_filename: Nombre del archivo donde se guardarán los centroides.
    """
    # Preprocesamiento de los datos (normalización)
    scaler = StandardScaler()
    X_2024 = scaler.fit_transform(df_2024)
    
    # Crear y entrenar el modelo KMeans
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    kmeans.fit(X_2024)
    
    # Obtener los centroides
    centroids = kmeans.cluster_centers_
    
    # Guardar el modelo (solo los centroides y el scaler)
    with open(model_filename, 'wb') as f:
        pickle.dump({'centroids': centroids, 'scaler': scaler}, f)
    
    print(f"Modelo entrenado y centroides guardados en {model_filename}")
    return centroids
```

```python
# Función para cargar el modelo y predecir el segmento de los nuevos clientes
def predict_new_customers(df_new, model_filename="kmeans_model.pkl"):
    """
    Predice el segmento (cluster) de los nuevos clientes utilizando los centroides guardados.
    df_new: DataFrame con los nuevos datos de clientes.
    model_filename: Nombre del archivo con el modelo (centroides y scaler).
    """
    # Cargar el modelo previamente entrenado
    with open(model_filename, 'rb') as f:
        model_data = pickle.load(f)
        centroids = model_data['centroids']
        scaler = model_data['scaler']
    
    # Preprocesar los datos de los nuevos clientes (normalización)
    X_new = scaler.transform(df_new)
    
    # Calcular las distancias entre cada nuevo cliente y los centroides
    distances = np.linalg.norm(X_new[:, np.newaxis] - centroids, axis=2)
    
    # Asignar el cluster más cercano a cada nuevo cliente
    labels_new = np.argmin(distances, axis=1)
    
    return labels_new
```
